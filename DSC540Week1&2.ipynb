#### Get the type of multiline data

```python
type(multiline_text)
```

#### Get the length of the multiline text

#### Cleaning the multiline text statement

```python
multiline_text = multiline_text.replace('\n', "")

multiline_text
```

#### Removing special chars and punctuation and other unwanted characters from text

```python

cleaned_multiline_text = ""
for char in multiline_text:
    if char == " ":
        cleaned_multiline_text += char
    elif char.isalnum():  # using the isalnum() method of strings.
        cleaned_multiline_text += char
    else:
        cleaned_multiline_text += " "

cleaned_multiline_text
```

#### Making list from the words in the text

```python
list_of_words = cleaned_multiline_text.split()
list_of_words
```

#### Getting Total length of the list

```python
len(list_of_words)
```

#### Now figure out unique word count from the list

```python
unique_words_as_dict = dict.fromkeys(list_of_words)
len(list(unique_words_as_dict.keys()))
```

```python
for word in list_of_words:
    if unique_words_as_dict[word] is None:
        unique_words_as_dict[word] = 1
    else:
        unique_words_as_dict[word] += 1
unique_words_as_dict
```

#### Now it is interesting and let's get Top 25 words from the unique word list 

```python
top_words = sorted(unique_words_as_dict.items(), key=lambda key_val_tuple: key_val_tuple[1], reverse=True)
top_words[:25]
```

```python
from itertools import permutations, dropwhile
permutations?
dropwhile?
```

#### This function generates all possible arrangements (permutations) of elements from an iterable. 
#### If r is specified then it generates permutations of length r; 
#### otherwise it generates permutations of the length of the iterable.

```python
permutations(range(3))
```

#### The isinstance() function in Python checks if an object belongs to a specific type or a tuple of types. 
#### If the object matches any type within the tuple, it returns True; otherwise, it returns False. 

```python
for number_tuple in permutations(range(3)):
    print(number_tuple)
    assert isinstance(number_tuple, tuple)  
```

```python
for number_tuple in permutations(range(3)):
    print(list(dropwhile(lambda x: x <= 0, number_tuple))) 
```

#### Convert to number from tuple list above created using permutations

```python
import math
def convert_to_number(number_stack):
    final_number = 0
    for i in range(0, len(number_stack)):
        final_number += (number_stack.pop() * (math.pow(10, i)))
    return final_number

for number_tuple in permutations(range(3)):
    number_stack = list(dropwhile(lambda x: x <= 0, number_tuple))
    print(convert_to_number(number_stack))
```

```python
from itertools import zip_longest  
```

```python
def return_dict_from_csv_line(header, line):
    # Zip them
    zipped_line = zip_longest(header, line, fillvalue=None)
    # Use dict comprehension to generate the final dict
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict  
```

### Reading the data from Sales_record CSV file and return the records until it reaches 10th row.

```python
with open("../DSC540Assignment/sales_record.csv", "r") as fd:
    first_line = fd.readline()
    header = first_line.replace("\n", "").split(",")
    for i, line in enumerate(fd):
        # Here we loop over the first 10 lines in order to not to make the output too big
        line = line.replace("\n", "").split(",")
        d = return_dict_from_csv_line(header, line)
        print(d)
        if i > 10:
            break
```
