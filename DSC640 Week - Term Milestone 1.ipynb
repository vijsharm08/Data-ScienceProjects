# National Database of Childcare Prices - Comprehensive Data Analysis

This notebook provides a detailed analysis of childcare prices across the United States, exploring variations by geography, age groups, and time periods.

## 1. Load and Explore the Dataset

Import necessary libraries and load the data to understand its structure.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
warnings.filterwarnings('ignore')

# Set visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Load the dataset
file_path = './nationaldatabaseofchildcareprices.xlsx'
df = pd.read_excel(file_path)

print(f"Dataset loaded successfully!")
print(f"Shape: {df.shape}")
```

```python
# Display first few rows
df.head(10)
```

```python
# Display column names and data types
print("Column Names and Data Types:")
print("="*60)
df.dtypes
```

```python
# Get basic information about the dataset
df.info()
```

## 2. Data Cleaning and Preprocessing

Check for data quality issues and prepare the dataset for analysis.

```python
# Check for missing values
missing_data = df.isnull().sum()
missing_percent = (missing_data / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing_Count': missing_data,
    'Percentage': missing_percent
}).sort_values('Percentage', ascending=False)

print("Missing Values Summary:")
print("="*60)
print(missing_df[missing_df['Missing_Count'] > 0])
```

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Display basic statistics about the dataset
print(f"\nDataset Overview:")
print(f"Total rows: {len(df):,}")
print(f"Total columns: {len(df.columns)}")
```

```python
# Identify categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
print(f"Categorical columns ({len(categorical_columns)}):")
for col in categorical_columns:
    print(f"  - {col}")
```

```python
# Create a copy for analysis to preserve original data
df_clean = df.copy()

# Display unique values for key categorical columns (first few)
print("Unique values in key categorical columns:")
print("="*60)
for col in categorical_columns[:5]:
    n_unique = df_clean[col].nunique()
    print(f"\n{col}: {n_unique} unique values")
    if n_unique <= 10:
        print(df_clean[col].unique())
```

## 3. Statistical Summary and Descriptive Analysis

Generate summary statistics for numerical columns and analyze distributions.

```python
# Generate descriptive statistics for all numeric columns
df_clean.describe()
```

```python
# More detailed statistics including additional percentiles
df_clean.describe(percentiles=[.1, .25, .5, .75, .9, .95, .99])
```

```python
# Identify potential price-related columns
price_columns = [col for col in df_clean.columns if any(keyword in col.lower() 
                 for keyword in ['price', 'cost', 'fee', 'mhi', 'wage', 'income'])]

print(f"Price-related columns ({len(price_columns)}):")
for col in price_columns:
    print(f"  - {col}")
```

```python
# Summary statistics for price-related columns
if price_columns:
    price_stats = df_clean[price_columns].describe()
    print("Summary Statistics for Price-Related Columns:")
    print("="*60)
    print(price_stats)
```

```python
# Calculate additional statistics for price columns
if price_columns:
    for col in price_columns[:5]:  # First 5 price columns
        print(f"\n{col}:")
        print(f"  Mean: ${df_clean[col].mean():,.2f}")
        print(f"  Median: ${df_clean[col].median():,.2f}")
        print(f"  Std Dev: ${df_clean[col].std():,.2f}")
        print(f"  Min: ${df_clean[col].min():,.2f}")
        print(f"  Max: ${df_clean[col].max():,.2f}")
        print(f"  Non-null count: {df_clean[col].count():,}")
```

## 4. Geographic Analysis of Childcare Prices

Analyze childcare prices across different geographic regions.

```python
# Identify geographic columns
geo_columns = [col for col in df_clean.columns if any(keyword in col.lower() 
               for keyword in ['state', 'county', 'region', 'city'])]

print(f"Geographic columns ({len(geo_columns)}):")
for col in geo_columns:
    print(f"  - {col}")
```

```python
# Analyze by state (if state column exists)
state_col = next((col for col in df_clean.columns if 'state' in col.lower()), None)

if state_col:
    print(f"Number of unique states: {df_clean[state_col].nunique()}")
    print(f"\nStates in dataset:")
    print(df_clean[state_col].value_counts().head(20))
```

```python
# Calculate average prices by state for different childcare types
if state_col and price_columns:
    # Select a few key price columns for state-level analysis
    key_price_cols = price_columns[:5]
    
    state_prices = df_clean.groupby(state_col)[key_price_cols].mean().round(2)
    state_prices = state_prices.sort_values(by=key_price_cols[0], ascending=False)
    
    print("Average Prices by State (Top 15):")
    print("="*60)
    print(state_prices.head(15))
```

```python
# Visualize top 15 states by childcare price (using first price column)
if state_col and price_columns:
    plt.figure(figsize=(14, 8))
    
    state_avg = df_clean.groupby(state_col)[price_columns[0]].mean().sort_values(ascending=False).head(15)
    
    sns.barplot(x=state_avg.values, y=state_avg.index, palette='viridis')
    plt.xlabel('Average Price ($)', fontsize=12)
    plt.ylabel('State', fontsize=12)
    plt.title(f'Top 15 States by Average {price_columns[0]}', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
```

```python
# Visualize bottom 15 states by childcare price
if state_col and price_columns:
    plt.figure(figsize=(14, 8))
    
    state_avg_low = df_clean.groupby(state_col)[price_columns[0]].mean().sort_values(ascending=True).head(15)
    
    sns.barplot(x=state_avg_low.values, y=state_avg_low.index, palette='coolwarm')
    plt.xlabel('Average Price ($)', fontsize=12)
    plt.ylabel('State', fontsize=12)
    plt.title(f'Bottom 15 States by Average {price_columns[0]}', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
```

```python
# Geographic price variation analysis
if state_col and price_columns:
    geo_analysis = df_clean.groupby(state_col)[price_columns[0]].agg(['mean', 'median', 'std', 'min', 'max'])
    geo_analysis = geo_analysis.sort_values('mean', ascending=False)
    
    print("Geographic Price Variation (Top 10 States):")
    print("="*60)
    print(geo_analysis.head(10).round(2))
```

## 5. Price Comparison Across Different Age Groups

Compare childcare prices for different age categories.

```python
# Identify age-related price columns
age_price_columns = [col for col in price_columns if any(keyword in col.lower() 
                     for keyword in ['infant', 'toddler', 'preschool', 'school', 'age'])]

print(f"Age-related price columns ({len(age_price_columns)}):")
for col in age_price_columns:
    print(f"  - {col}")
```

```python
# Compare average prices across age groups
if age_price_columns:
    age_comparison = df_clean[age_price_columns].mean().sort_values(ascending=False)
    
    print("Average Prices by Age Group:")
    print("="*60)
    for col, price in age_comparison.items():
        print(f"{col}: ${price:,.2f}")
```

```python
# Visualize price comparison across age groups
if age_price_columns and len(age_price_columns) > 0:
    plt.figure(figsize=(12, 6))
    
    age_means = df_clean[age_price_columns].mean().sort_values(ascending=False)
    
    sns.barplot(x=age_means.index, y=age_means.values, palette='Set2')
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Average Price ($)', fontsize=12)
    plt.xlabel('Age Group Category', fontsize=12)
    plt.title('Average Childcare Prices by Age Group', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
```

```python
# Box plot comparison across age groups
if age_price_columns and len(age_price_columns) > 0:
    plt.figure(figsize=(14, 7))
    
    # Prepare data for box plot
    age_data = df_clean[age_price_columns].melt(var_name='Age_Group', value_name='Price')
    age_data = age_data.dropna()
    
    sns.boxplot(data=age_data, x='Age_Group', y='Price', palette='Set3')
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Price ($)', fontsize=12)
    plt.xlabel('Age Group', fontsize=12)
    plt.title('Distribution of Childcare Prices by Age Group', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
```

```python
# Statistical comparison of age groups
if age_price_columns:
    print("Detailed Statistics by Age Group:")
    print("="*60)
    age_stats = df_clean[age_price_columns].describe()
    print(age_stats.round(2))
```

## 6. Time Series Analysis of Price Trends

Analyze how childcare prices have changed over time (if temporal data is available).

```python
# Identify time-related columns
time_columns = [col for col in df_clean.columns if any(keyword in col.lower() 
                for keyword in ['year', 'date', 'time', 'period'])]

print(f"Time-related columns ({len(time_columns)}):")
for col in time_columns:
    print(f"  - {col}")
    print(f"  Unique values: {df_clean[col].nunique()}")
    if df_clean[col].nunique() <= 20:
        print(f"  Values: {sorted(df_clean[col].unique())}")
```

```python
# Time series analysis if year column exists
year_col = next((col for col in time_columns if 'year' in col.lower()), None)

if year_col and price_columns:
    yearly_trends = df_clean.groupby(year_col)[price_columns[:3]].mean()
    
    print("Average Prices by Year:")
    print("="*60)
    print(yearly_trends.round(2))
```

```python
# Visualize price trends over time
if year_col and price_columns:
    plt.figure(figsize=(14, 7))
    
    yearly_avg = df_clean.groupby(year_col)[price_columns[:5]].mean()
    
    for col in price_columns[:5]:
        plt.plot(yearly_avg.index, yearly_avg[col], marker='o', label=col, linewidth=2)
    
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Average Price ($)', fontsize=12)
    plt.title('Childcare Price Trends Over Time', fontsize=14, fontweight='bold')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
# Calculate year-over-year growth rates
if year_col and price_columns:
    yearly_prices = df_clean.groupby(year_col)[price_columns[0]].mean()
    yoy_growth = yearly_prices.pct_change() * 100
    
    print("Year-over-Year Growth Rates (%):")
    print("="*60)
    print(yoy_growth.round(2))
```

```python
# Visualize year-over-year growth
if year_col and price_columns:
    plt.figure(figsize=(12, 6))
    
    yearly_prices = df_clean.groupby(year_col)[price_columns[0]].mean()
    yoy_growth = yearly_prices.pct_change() * 100
    
    plt.bar(yoy_growth.index, yoy_growth.values, color=['green' if x > 0 else 'red' for x in yoy_growth.values])
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Growth Rate (%)', fontsize=12)
    plt.title(f'Year-over-Year Growth Rate: {price_columns[0]}', fontsize=14, fontweight='bold')
    plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```

## 7. Correlation Analysis

Examine relationships between childcare prices and other variables.

```python
# Calculate correlation matrix for numeric columns
correlation_matrix = df_clean[numeric_columns].corr()

print("Correlation matrix shape:", correlation_matrix.shape)
```

```python
# Display correlations with price columns
if price_columns:
    price_correlations = correlation_matrix[price_columns[0]].sort_values(ascending=False)
    
    print(f"Correlations with {price_columns[0]}:")
    print("="*60)
    print(price_correlations.head(20))
```

```python
# Visualize correlation heatmap for price-related columns
if len(price_columns) > 1:
    plt.figure(figsize=(14, 10))
    
    # Select price columns and highly correlated variables
    cols_for_heatmap = price_columns[:10] if len(price_columns) > 10 else price_columns
    corr_subset = df_clean[cols_for_heatmap].corr()
    
    sns.heatmap(corr_subset, annot=True, fmt='.2f', cmap='coolwarm', 
                center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})
    plt.title('Correlation Heatmap: Price-Related Variables', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
```

```python
# Full correlation heatmap (for all numeric variables)
plt.figure(figsize=(20, 16))

sns.heatmap(correlation_matrix, cmap='coolwarm', center=0, 
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Full Correlation Heatmap: All Numeric Variables', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Identify strong correlations (excluding self-correlation)
strong_correlations = []

for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_value = correlation_matrix.iloc[i, j]
        if abs(corr_value) > 0.7:  # Strong correlation threshold
            strong_correlations.append({
                'Variable 1': correlation_matrix.columns[i],
                'Variable 2': correlation_matrix.columns[j],
                'Correlation': corr_value
            })

strong_corr_df = pd.DataFrame(strong_correlations).sort_values('Correlation', 
                                                                 key=abs, ascending=False)

print(f"Strong Correlations (|r| > 0.7): {len(strong_corr_df)}")
print("="*60)
print(strong_corr_df.head(20))
```

```python
# Scatter plot for highly correlated variables
if len(strong_corr_df) > 0 and len(strong_corr_df) > 0:
    fig, axes = plt.subplots(2, 2, figsize=(14, 12))
    axes = axes.ravel()
    
    for idx, row in strong_corr_df.head(4).iterrows():
        var1, var2 = row['Variable 1'], row['Variable 2']
        
        ax = axes[idx if idx < 4 else 3]
        ax.scatter(df_clean[var1], df_clean[var2], alpha=0.5, s=10)
        ax.set_xlabel(var1, fontsize=10)
        ax.set_ylabel(var2, fontsize=10)
        ax.set_title(f'r = {row["Correlation"]:.3f}', fontsize=11)
        ax.grid(True, alpha=0.3)
    
    plt.suptitle('Scatter Plots: Highly Correlated Variables', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
```

## 8. Visualization of Key Findings

Comprehensive visualizations to communicate main insights.

```python
# Create a comprehensive dashboard visualization
if price_columns and state_col:
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    
    # 1. Distribution of main price variable
    axes[0, 0].hist(df_clean[price_columns[0]].dropna(), bins=50, color='skyblue', edgecolor='black')
    axes[0, 0].set_xlabel('Price ($)', fontsize=11)
    axes[0, 0].set_ylabel('Frequency', fontsize=11)
    axes[0, 0].set_title(f'Distribution of {price_columns[0]}', fontsize=12, fontweight='bold')
    axes[0, 0].grid(True, alpha=0.3)
    
    # 2. Top 10 states by average price
    state_avg = df_clean.groupby(state_col)[price_columns[0]].mean().sort_values(ascending=False).head(10)
    axes[0, 1].barh(range(len(state_avg)), state_avg.values, color='coral')
    axes[0, 1].set_yticks(range(len(state_avg)))
    axes[0, 1].set_yticklabels(state_avg.index)
    axes[0, 1].set_xlabel('Average Price ($)', fontsize=11)
    axes[0, 1].set_title('Top 10 States by Average Price', fontsize=12, fontweight='bold')
    axes[0, 1].grid(True, alpha=0.3, axis='x')
    
    # 3. Price comparison across age groups (if available)
    if age_price_columns and len(age_price_columns) > 0:
        age_means = df_clean[age_price_columns[:5]].mean().sort_values(ascending=False)
        axes[1, 0].bar(range(len(age_means)), age_means.values, color='lightgreen', edgecolor='black')
        axes[1, 0].set_xticks(range(len(age_means)))
        axes[1, 0].set_xticklabels([col[:20] for col in age_means.index], rotation=45, ha='right', fontsize=9)
        axes[1, 0].set_ylabel('Average Price ($)', fontsize=11)
        axes[1, 0].set_title('Average Prices by Age Group', fontsize=12, fontweight='bold')
        axes[1, 0].grid(True, alpha=0.3, axis='y')
    
    # 4. Box plot of price distribution
    price_data = df_clean[price_columns[:5]].melt(var_name='Category', value_name='Price')
    price_data = price_data.dropna()
    
    categories = price_data['Category'].unique()[:5]
    box_data = [price_data[price_data['Category'] == cat]['Price'].values for cat in categories]
    
    axes[1, 1].boxplot(box_data, labels=[cat[:15] for cat in categories])
    axes[1, 1].set_xticklabels([cat[:15] for cat in categories], rotation=45, ha='right', fontsize=9)
    axes[1, 1].set_ylabel('Price ($)', fontsize=11)
    axes[1, 1].set_title('Price Distribution by Category', fontsize=12, fontweight='bold')
    axes[1, 1].grid(True, alpha=0.3, axis='y')
    
    plt.suptitle('Childcare Prices: Key Insights Dashboard', fontsize=16, fontweight='bold', y=1.00)
    plt.tight_layout()
    plt.show()
```

```python
# Summary statistics visualization
if price_columns:
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    # Price range comparison
    price_ranges = df_clean[price_columns[:8]].agg(['min', 'max', 'mean'])
    x_pos = np.arange(len(price_ranges.columns))
    
    axes[0].bar(x_pos - 0.2, price_ranges.loc['min'], width=0.2, label='Min', color='lightblue')
    axes[0].bar(x_pos, price_ranges.loc['mean'], width=0.2, label='Mean', color='orange')
    axes[0].bar(x_pos + 0.2, price_ranges.loc['max'], width=0.2, label='Max', color='lightcoral')
    axes[0].set_xticks(x_pos)
    axes[0].set_xticklabels([col[:20] for col in price_ranges.columns], rotation=45, ha='right', fontsize=9)
    axes[0].set_ylabel('Price ($)', fontsize=11)
    axes[0].set_title('Price Range Comparison', fontsize=12, fontweight='bold')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3, axis='y')
    
    # Coefficient of variation (to show relative variability)
    cv_data = (df_clean[price_columns[:8]].std() / df_clean[price_columns[:8]].mean()) * 100
    cv_data = cv_data.sort_values(ascending=False)
    
    axes[1].barh(range(len(cv_data)), cv_data.values, color='mediumpurple')
    axes[1].set_yticks(range(len(cv_data)))
    axes[1].set_yticklabels([col[:25] for col in cv_data.index], fontsize=9)
    axes[1].set_xlabel('Coefficient of Variation (%)', fontsize=11)
    axes[1].set_title('Price Variability Across Categories', fontsize=12, fontweight='bold')
    axes[1].grid(True, alpha=0.3, axis='x')
    
    plt.tight_layout()
    plt.show()
```

```python
# Geographic heatmap-style visualization
if state_col and price_columns:
    state_summary = df_clean.groupby(state_col)[price_columns[:5]].mean()
    
    plt.figure(figsize=(14, 10))
    sns.heatmap(state_summary, annot=False, cmap='YlOrRd', cbar_kws={'label': 'Average Price ($)'})
    plt.title('Childcare Prices Heatmap by State and Category', fontsize=14, fontweight='bold')
    plt.xlabel('Price Category', fontsize=12)
    plt.ylabel('State', fontsize=12)
    plt.tight_layout()
    plt.show()
```
