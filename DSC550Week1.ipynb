## Five publishers with the highest total global sales

```python
# Group by publisher and sum the global sales
publisher_sales = videogamedf.groupby('Publisher')['Global_Sales'].sum()

# Sort in descending order and get the top 5
top_publishers = publisher_sales.sort_values(ascending=False).head(5)

# Print the results
print(top_publishers)
```

## Calculate the percentage of North American sales

```python
# Calculate the percentage of North American sales
videogamedf['Percentage of Global Sales'] = (videogamedf['NA_Sales'] / videogamedf['Global_Sales']) * 100

# Display the first five rows
df = videogamedf[['Name','Platform','Year_of_Release','Genre','NA_Sales','Global_Sales','Percentage of Global Sales']]

print(df.head(5))
```

## Median vlaues to fill in for 'NaN' values

```python
user_scores = videogamedf['User_Score']
non_numeric_string = 'tbd'
user_scores = user_scores.replace(non_numeric_string, pd.NA)

# Convert the user score column to numeric
user_scores = pd.to_numeric(user_scores)

# Calculate the median user score
median_user_score = user_scores.median()

# Replace NaN values with the median
user_scores = user_scores.fillna(median_user_score)

# Update the 'user_score' column in the DataFrame
videogamedf['User_Score'] = user_scores
```
